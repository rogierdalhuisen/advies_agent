```python
import os
import requests
from dotenv import load_dotenv
import json

# Laad je .env bestand
load_dotenv()

# Haal de endpoints en token op
RELATIES_URL = os.getenv('ASSUPORTAL_RELATIES')
CONTRACTEN_URL = os.getenv('ASSUPORTAL_CONTRACTEN')
BEARER_TOKEN = os.getenv('ASSUPORTAL_API_TOKEN')  # Of hoe je het token ook genoemd hebt

GRIP_ENDPOINT = os.getenv('GRIP_ENDPOINT')
GRIP_BEARER_TOKEN = os.getenv('GRIP_BEARER_TOKEN')

print(f"e-grip Endpoint: {GRIP_ENDPOINT}")
print(f"Token aanwezig: {'Ja' if GRIP_BEARER_TOKEN else 'Nee'}")

print(f"Relaties URL: {RELATIES_URL}")
print(f"Contracten URL: {CONTRACTEN_URL}")
print(f"Token aanwezig: {'Ja' if BEARER_TOKEN else 'Nee'}")
```

```python
form_id = 2  # Start with the example ID from the email

grip_headers = {
    'Authorization': f'Bearer {GRIP_BEARER_TOKEN}'
}

try:
    print(f"Making API call to e-grip for formId={form_id}...")
    response_grip = requests.get(
        f"{GRIP_ENDPOINT}?formId={form_id}",
        headers=grip_headers,
        timeout=30
    )

    response_grip.raise_for_status()

    print(f"✓ Success! Status code: {response_grip.status_code}")

    data_grip = response_grip.json()

    print("\n--- Complete JSON Response ---")
    print(json.dumps(data_grip, indent=2, ensure_ascii=False))

except requests.exceptions.RequestException as e:
    print(f"✗ Error: {e}")
    if hasattr(e, 'response') and e.response is not None:
        print(f"Response text: {e.response.text}")
```
